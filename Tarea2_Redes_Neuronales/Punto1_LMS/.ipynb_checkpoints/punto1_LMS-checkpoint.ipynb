```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from pydoc import help  # can type in the python console `help(name of function)` to get the documentation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from scipy import stats
from IPython.display import display, HTML
from pandas.tools.plotting import scatter_matrix
from matplotlib.artist import setp
import time

plt.rcParams['figure.figsize'] = (10,16)
```

```python
df1     = pd.io.excel.read_excel("DatosPunto1.xlsx", sheetname=0)
classes = df1.iloc[:,-1].copy()
df1 = df1.iloc[:,0:-2].copy()
#normalización min-max
df1 = (df1-np.min(df1, axis = 0))/(np.max(df1, axis = 0)-np.min(df1, axis = 0))

y = np.array(classes)
X = np.array(df1)
#random values
i = np.random.permutation(X.shape[0])
X = X[i,:]
y = y[i]
```

```python
def derivating(fx):
    difcoefs = [1,-1]
    return np.convolve(fx,difcoefs,mode='same')

def LMS_training(X,y,iteraciones=500,alpha=0.001,de_min=0.01):
    
    ##falta mirar como imprimir Error en tiempo real
    
    w_          = np.random.rand(X.shape[1] + 1,)
    n           = X.shape[0]
    nf          = X.shape[1]
    sw          = 0
    j           = 0
    index       = []
    error_array = []
    

    while j < iteraciones and sw == 0:
        for i in range(0,n):
            g  = np.dot(w_,np.insert(X[i,:],0,1).T)
            e  = g - y[i]
            w_ = w_ - alpha*e*np.insert(X[i,:],0,1)
        predicted = []
        for i in range(0,n):
            predicted.append(Predict(X[i,:],w_))
        g = (y-predicted)
        E = 0.5*np.sum(g*g.T)/y.shape #queda la duda
        
        index.append(j)
        error_array.append(E[0])
        d_error_array = derivating(error_array)
        
        if np.abs(d_error_array[len(d_error_array)-1]) < de_min:
            sw = 1
        
        j = j + 1
    print('Función de error: '+str(E[0])+', iteraciones: '+str(j))
    plt.plot(index,error_array,'r')
    plt.xlabel('Iteraciones')
    plt.ylabel('Función de error')
    plt.grid()
    plt.show()
    plt.plot(index,d_error_array)
    plt.xlabel('Iteraciones')
    plt.ylabel('derivada de la Función de error')
    plt.grid()
    plt.show()
    
    return [w_,predicted]

def Predict(X,w_): 
    return np.dot(w_,np.insert(X,0,1).T)
```

### Entrenamiento de la red neuronal usando validación bootstraping

```python
N_samples = X.shape[0]
N_train   = int(0.7*N_samples)
N_val     = N_samples-N_train

X_train = X[0:N_train,:]
X_val   = X[N_train:N_samples,:]

y_train = y[0:N_train]
y_val   = y[N_train:N_samples]

[wLMS,predicted] = LMS_training(X_train,y_train,iteraciones=1000,de_min=0.1)
```

```python
d = {'datos_predichos': predicted, 'diferencia' : np.abs(predicted-y_train),'etiquetas_entrenamiento': y_train}
df = pd.DataFrame(data=d)
df
```

```python
ecm1 = np.sqrt(np.sum(((predicted-y_train)*(predicted-y_train)))/y_train.shape)[0]
print('error cuadrático medio de entrenamiento: '+str(ecm1))
```

```python
predicted2 = []
for i in range(0,X_val.shape[0]):
    predicted2.append(Predict(X_val[i,:],wLMS))

ecm2 = np.sqrt(np.sum(((predicted2-y_val)*(predicted2-y_val)))/y_val.shape)[0]
print('error cuadrático medio de validación: '+str(ecm2))
```
